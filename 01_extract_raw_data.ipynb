## 1.1解压压缩包

将 7 个 文件保留在 。.zipdata/data_raw_zip/
使用 Python 代码自动解压，不得手动逐个解压。
解压后的原始数据存放在 。data/raw/
在 中列出 7 个原始文件的名称、主要内容和你从中提取的变量。README.md

````python
# ==============================================
# 作业：01_extract_raw_data.ipynb
# 功能：批量解压7个文件 + 自动识别Excel/CSV + 读取主数据
# ==============================================
import pandas as pd
import zipfile
import os
import warnings
warnings.filterwarnings('ignore')

# ===================== 固定路径 =====================
zip_folder = "data/raw/CSMAR/data_raw_zip/"

# 1. 批量解压所有7个压缩包
print("🔄 开始批量解压所有CSMAR压缩包...")
for file_name in os.listdir(zip_folder):
    if file_name.endswith(".zip"):
        zip_path = os.path.join(zip_folder, file_name)
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(zip_folder)
        print(f"✅ 已解压：{file_name}")
print("\n🎉 全部7个压缩包 自动解压完成！")

# ===================== 扫描所有数据文件（CSV + Excel） =====================
print("\n📂 所有解压后的 数据文件：")
all_files = []
for root, dirs, files in os.walk(zip_folder):
    for file in files:
        # 同时识别 CSV 和 Excel 文件
        if file.endswith((".csv", ".xlsx", ".xls")):
            full_path = os.path.join(root, file)
            all_files.append(full_path)
            print(f"- {full_path}")

# ===================== 自动读取「常用变量」主数据 =====================
print("\n🔄 正在读取 CSMAR 常用变量 主数据...")
df = None
for file_path in all_files:
    if "常用变量" in file_path:
        # 自动判断文件格式
        if file_path.endswith(".csv"):
            df = pd.read_csv(file_path, encoding='gbk')
        else:
            df = pd.read_excel(file_path)
        print(f"✅ 成功读取：{os.path.basename(file_path)}")
        break

# ===================== 原始数据初步识别（作业要求） =====================
print("\n" + "="*60)
print("📊 CSMAR 原始数据初步识别")
print("="*60)
print(f"数据规模：{df.shape[0]} 行 × {df.shape[1]} 列")
print("\n🔍 数据列名：")
print(df.columns.tolist())
print("\n🔍 数据基础信息：")
df.info()
print("\n🔍 前5行数据预览：")
print(df.head())
````

````python
# ==============================================
# 01_extract_raw_data.ipynb
# 功能：解压 CSMAR 原始数据 + 读取并初步识别
# ==============================================
import zipfile
import os

# 原始压缩包所在目录
zip_dir = "data/raw/CSMAR/data_raw_zip"
extract_to = "data/raw/CSMAR"

# 创建解压目标目录
os.makedirs(extract_to, exist_ok=True)

# 查找所有 zip 文件
zip_files = [f for f in os.listdir(zip_dir) if f.endswith(".zip")]
print(f"找到 {len(zip_files)} 个压缩包：")
for f in zip_files:
    print(f"  - {f}")

# 解压
for zf in zip_files:
    zip_path = os.path.join(zip_dir, zf)
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(extract_to)
    print(f"  已解压：{zf}")

print("\n解压完成！文件位于：", os.path.abspath(extract_to))
````
